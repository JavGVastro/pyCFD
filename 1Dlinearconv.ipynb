```python
import numpy as np
from matplotlib import pyplot
import time,sys
```

```python
%matplotlib inline
```

nx=number of grid points

dx=distance between adjacent grid points

```python
nx=41
dx=2/(nx-1)
nt=25
dt=.025
c=1
```

u0=u=2
0.5<=x<=1 and u=1 everuwhere between 0,2

```python
u=np.ones(nx)
u[int(.5/dx):int(1/dx+1)]=2
print(u)
```

```python
pyplot.plot(np.linspace(0,2,nx),u);
```

```python
un=np.ones(nx)
for n in range(nt):
    un=u.copy()
    for i in range(1,nx):
        u[i]=un[i]-c*dt/dx*(un[i]-un[i-1])
```

```python
pyplot.plot(np.linspace(0,2,nx),u);
```
